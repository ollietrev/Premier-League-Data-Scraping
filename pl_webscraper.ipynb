Web scraping premier league data to predict 24/25 winners

```python
import pandas as pd # Importing necessary libraries for web scraping
import requests
from bs4 import BeautifulSoup
```

```python
url_start = "https://fbref.com/en/comps/9/{}-{}/{}-{}-Premier-League-Stats"

years = list(range(2009, 2024))

        
```

```python
from selenium import webdriver
import time
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)

for year in years:
    url = url_start.format(year, year+1, year, year+1)
    driver.get(url)
    driver.execute_script("window.scrollTo(1, 10000)")
    time.sleep(2)

    html = driver.page_source
    with open("plstats/{}.html".format(year), "w+") as f:
        f.write(html)
```

```python
df = []
for year in years:
    with open("plstats/{}.html".format(year), "r") as f:
        page = f.read()
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="over_header").decompose()
    stats_table = soup.find(id="all_stats_squads_standard")
    stats = pd.read_html(str(stats_table))[0]
    stats["Year"] = year
    df.append(stats)
    
```

```python
teams = pd.concat(df)
teams
```
